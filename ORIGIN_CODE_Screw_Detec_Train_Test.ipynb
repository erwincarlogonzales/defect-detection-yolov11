- Precision and Recall slightly improved, showing better identification and coverage of instances.

**Class-wise Performance:**

- The rivet class saw the largest improvement (+0.006), which suggests 100 epochs helped refine harder-to-classify objects.
- Minimal improvements for other classes, showing diminishing returns after 50 epochs.

**Loss Analysis:**

- Both training and validation losses are consistently low at 100 epochs, indicating no overfitting.
- The learning curve is stable, meaning the additional epochs focused on fine-tuning.

**Confusion Matrix:**

- Both confusion matrices are highly diagonal for 50 and 100 epochs, with minimal misclassifications.
- The additional training reduced minor errors further in rare classes like rivet and washer.

**Conclusion:**

- 100 epochs slightly outperforms 50 epochs, with marginal improvements in metrics.
- No signs of overfitting, as validation loss continues to decrease and metrics improve.
